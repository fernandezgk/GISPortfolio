```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('county_election_data_2000-2016.csv')
df.head()
```

```python
a = abs(df.gop_minus_dem_prc_2008)
b = abs(df.gop_minus_dem_prc_2012)
c = abs(df.gop_minus_dem_prc_2016)

a2 = np.sum(a>=20) #2008landslide
b2 = np.sum(b>=20) #2012landslide
c2 = np.sum(c>=20) #20162landslide

leng = len(df.FIPS)

a3 = (a2/leng)*100
b3 = (b2/leng)*100
c3 = (c2/leng)*100
print()

d = np.sum(a<=10) #2008Competitive
e = np.sum(b<=10) #2012Competitive
f = np.sum(c<=10) #2016Competitive
print()

d2 = (d/leng)*100
e2 = (e/leng)*100
f2 = (f/leng)*100
print()

plt.plot([a3,b3,c3], label = 'Landslide Counties', color = 'orange')
plt.plot([d2,e2,f2], label = 'Competitive Counties', color = 'maroon')
plt.xlabel('Year')
plt.ylabel('County Percentage')
plt.title('Presidential Election Landslide & Competative Counties')
plt.xticks(np.arange(3), ['2008','2012','2016'])
plt.legend()
plt.savefig('lsncc.png', dpi = 300)
```
